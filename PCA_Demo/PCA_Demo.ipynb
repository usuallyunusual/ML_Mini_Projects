# Demonstrating PCA

## Importing necessary libraries
Get the dataset at : https://www.kaggle.com/tongpython/cat-and-dog

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
import seaborn as sns; sns.set()
import glob
from sklearn.utils import shuffle
from PIL import Image
import os, sys
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import PCA
```

## Resizing all of the images in the dataset to 64 x 64
### Data consists of 10000 images of cats and dogs that are of different sizes.

```python
paths = ["D:\\Education\\MSc\\Sem 3\\lab\\ML\\ML_prog4\\training_set\\cats\\",
         "D:\\Education\\MSc\\Sem 3\\lab\\ML\\ML_prog4\\training_set\\dogs\\",
         "D:\\Education\\MSc\\Sem 3\\lab\\ML\\ML_prog4\\test_set\\cats\\",
         "D:\\Education\\MSc\\Sem 3\\lab\\ML\\ML_prog4\\test_set\\dogs\\"]
def resize():
    for path in paths:
        dirs = os.listdir( path )

        for item in dirs:
            if os.path.isfile(path+item):
                im = Image.open(path+item)
                f, e = os.path.splitext(path+item)
                imResize = im.resize((32,32), Image.ANTIALIAS)
                imResize.save(f+'.jpg', 'JPEG', quality=90)

resize()
```

## Preparing training data
### Using glob to list files ( We are going to read only half the data i.e 4000 images for train set)**
** Due to memory constraints


```python
flist_cat = glob.glob("training_set/cats/*.jpg")
flist_cat = flist_cat[:int(len(flist_cat)/2)]
print(len(flist_cat))
flist_dog = glob.glob("training_set/dogs/*.jpg")
flist_dog = flist_dog[:int(len(flist_dog)/2)]
print(len(flist_dog))
```

### Read images (Using matlplotlib.image) into a numpy array
Image sizes are displayed

```python
cats = [image.imread(i) for i in flist_cat if image.imread(i).shape[0] == 32 and image.imread(i).shape[1] == 32]
dogs = [image.imread(i) for i in flist_dog if image.imread(i).shape[0] == 32 and image.imread(i).shape[1] == 32]
print(len(cats))
print(len(dogs))
train_set_x = np.array(cats+dogs)
print(type(train_set_x))
print(train_set_x.shape)
```

### Preparing a target array for the train set.
### Finally we have train_set_x ( train set data) and train_set_y ( targets )
Size of arrays are displayed

```python
cats_y = np.ones([1,len(cats)])
dogs_y = np.zeros([1,len(dogs)])
print(cats_y.shape)
print(dogs_y.shape)
train_set_y = np.concatenate((cats_y,dogs_y),axis = 1).T
print(train_set_y.shape)
```

## Preparing test set
### Using glob to list files ( We are reading about half the images i.e 500 )**
** Due to memory constraints

```python
flist_cat_test = glob.glob("test_set/cats/*.jpg")
flist_cat_test = flist_cat_test[:int(len(flist_cat_test)/2)]
print(len(flist_cat_test))
flist_dog_test = glob.glob("test_set/dogs/*.jpg")
flist_dog_test = flist_dog_test[:int(len(flist_dog_test)/2)]
print(len(flist_dog_test))
```

### Reading the images into numpy array
Size of images are displayed

```python
cats_test = [image.imread(i) for i in flist_cat_test if image.imread(i).shape[0] == 32 and image.imread(i).shape[1] == 32]
dogs_test = [image.imread(i) for i in flist_dog_test if image.imread(i).shape[0] == 32 and image.imread(i).shape[1] == 32]
print(len(cats_test))
print(len(dogs_test))
test_set_x = np.array(cats_test+dogs_test)
print(type(test_set_x))
print(test_set_x.shape)
```

### Preparing targets for test set
### Finally we have test_set_x ( test set data) and test_set_y ( targets )

Size of array is displayed

```python
cats_y_test = np.ones([1,len(cats_test)])
dogs_y_test = np.zeros([1,len(dogs_test)])
print(cats_y_test.shape)
print(dogs_y_test.shape)
test_set_y = np.concatenate((cats_y_test,dogs_y_test),axis = 1).T
print(test_set_y.shape)
```

## Shuffling the dataset using sklearn's shuffle ( )
### The sizes after shuffling are displayed


```python
train_set_x,train_set_y = shuffle(train_set_x,train_set_y)
test_set_x,test_set_y = shuffle(test_set_x,test_set_y)
```

```python
print(train_set_x.shape)
print(train_set_y.shape)
print(test_set_x.shape)
print(test_set_y.shape)
```

## Making a single vector out of a (64,64,3) image.
### Resultant sizes are displayed
#### PCA cannot handle more than 2 dimensional arrays. 

```python
train_vector = train_set_x.reshape(train_set_x.shape[0],-1)
test_vector = test_set_x.reshape(test_set_x.shape[0],-1)
print(train_vector.shape)
print(test_vector.shape)
```

## PCA for visualization of data
### Using PCA ( ) to reduce dimension from 3072 to 2 ( 2D plot )

```python
pca = PCA(2)  
projected = pca.fit_transform(train_vector)
print(train_vector.shape)
print(projected.shape)
print("%0.2f"%(np.cumsum(pca.explained_variance_ratio_)[-1]*100),"% of variance is preserved")
```

```python
plt.scatter(projected[:, 0], projected[:, 1],
            c=train_set_y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('brg', 2))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

### Using PCA ( ) to reduce dimesnionality from 3072 to 3 (3D plot)

```python
pca = PCA(3)  
projected = pca.fit_transform(train_vector)
print(train_set_x.shape)
print(projected.shape)
print("%0.2f"%(np.cumsum(pca.explained_variance_ratio_)[-1]*100),"% of variance is preserved")
```

```python
ax = Axes3D(plt.figure())
p = ax.scatter(projected[:, 0], projected[:, 1],projected[:,2],
            c=train_set_y, alpha=0.5,
            cmap=plt.cm.get_cmap('brg', 2))
plt.xlabel('component 1')
plt.ylabel('component 2')
ax.set_zlabel('component 3')
plt.colorbar(p);
```

## PCA ( ) to reduce dimensionality using number of components

```python
pca = PCA(10)  # project from 64 to 2 dimensions
projected = pca.fit_transform(train_vector)
print(train_set_x.shape)
print(projected.shape)
print("%0.2f"%(np.cumsum(pca.explained_variance_ratio_)[-1]*100),"% of variance is preserved")
```

## PCA ( ) to reduce dimensionality using percentage of explained variance

```python
pca = PCA(0.9)  
projected = pca.fit_transform(train_vector)
print(train_set_x.shape)
print(projected.shape)
print("%0.2f"%(np.cumsum(pca.explained_variance_ratio_)[-1]*100),"% of variance is preserved")
```

## Visualizing the number of dimensions vs the preserved variance plot
### We can see that for about 145 dimensions we get 90% variance

```python
pca = PCA().fit(train_vector)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
cumsums = np.cumsum(pca.explained_variance_ratio_)
print(cumsums[80]*100)
```

## Training a logistic regression model on original data set
### Takes about 10 minutes to train 
### Accuracy is reported at the end

```python
model = LogisticRegression(solver='liblinear', random_state=0).fit(train_vector,train_set_y)
```

```python
print("Test accuracy: %0.2f "%(model.score(test_vector,test_set_y)*100),"%")
print("Train accuracy: %0.2f "%(model.score(train_vector,train_set_y)*100),"%")
```

## Training logistic regression model on data after PCA ( ) is apllied
### Using PCA ( ) to reduce dimensionality

```python
pca = PCA(300)  # project from 64 to 2 dimensions
projected_x = pca.fit_transform(train_vector)
projected_y = pca.fit_transform(test_vector)
#recon = pca.inverse_transform(projected)
print(train_vector.shape)
print(projected_x.shape)
print("%0.2f"%(np.cumsum(pca.explained_variance_ratio_)[-1]*100),"% of variance is preserved")
```

### Model takes a few seconds to train ( 1500% decrease in time taken or almost 150 times faster)
### Accuracy is reported at the end

```python
model2 = LogisticRegression(solver='liblinear', random_state=0).fit(projected_x,train_set_y)
```

```python
print("Test accuracy: %0.2f "%(model2.score(projected_y,test_set_y)*100),"%")
print("Train accuracy: %0.2f "%(model2.score(projected_x,train_set_y)*100),"%")
```
