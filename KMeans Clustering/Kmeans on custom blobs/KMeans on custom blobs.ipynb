##  Demonstrate K Means Algorithm.
* Use at least one plot to visualize the results.

## Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
```

## Making the blobs. This is a function that creates samples for us
### We have 6 centres and 200 samples from this

```python
X,y = make_blobs(centers=6, n_samples=200, random_state=0, cluster_std=0.7)
print(X[:10],y[:10])
```

## Displaying a scatter to visualize the blob samples

```python
plt.scatter(X[:,0],X[:,1]);
```

## Using Sklearn KMeans function

Here we are using 6 centers and fitting the model to the blobs we have created. And we are printing the assigned clusters

```python
model = KMeans(6)
model.fit(X)
print(model.cluster_centers_)
```

## Visualizing the clustered data again using the same scatter plot but with colors

We are visualizing the centres of the clusters also

```python
plt.scatter(X[:,0],X[:,1], c=model.labels_);
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=100, color="red"); # Show the centres
```
